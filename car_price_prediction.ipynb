multicollinearity

```python
data_final.columns.values
```

```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
variable=data_final[['Mileage','EngineV','Year']]
vif=pd.DataFrame()

vif["VIF"] = [variance_inflation_factor(variable.values, i) for i in range(variable.shape[1])]
vif["features"] = variable.columns
```

```python
vif
```

```python
data_final=data_final.drop(['Year'],axis=1)
```

create dummy variable..

```python
d_data=pd.get_dummies(data_final,drop_first=True)
```

```python
d_data
```

```python
d_data.columns.values
```

```python
cols=['log_price','Mileage', 'EngineV',  'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_pre=d_data[cols]
data_pre.head()
```

Linear regression

```python
target=data_pre['log_price']
inputs=data_pre.drop(['log_price'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(inputs)
```

```python
data_scaler=scaler.transform(inputs)
```

```python
data_scaler
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2,random_state=365)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
pre_val=reg.predict(x_train)
```

```python
plt.scatter(y_train,pre_val)
plt.xlabel("Target: (Y_train)",size=18)
plt.ylabel("preditions: (pre_val)",size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train-pre_val)
plt.title("residual",size=15)
```

```python
reg.score(x_train,y_train)
```

finding weights and bias

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_sum=pd.DataFrame(inputs.columns.values,columns=['features'])
reg_sum['weights']=reg.coef_
reg_sum
```

Testing of the model

```python
y_pre_test=reg.predict(x_test)
```

```python
plt.scatter(y_test,y_pre_test,alpha=0.1)
plt.xlabel("Target: (y_test)",size=18)
plt.ylabel("preditions: (y_pre)",size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pre=pd.DataFrame(np.exp(y_pre_test),columns=['preditions'])
df_pre.head(5)
```

```python
y_test=y_test.reset_index(drop=True)
y_test
```

```python
df_pre['Target']=np.exp(y_test)
df_pre.head()
```

```python
df_pre['Residual']=df_pre['Target']-df_pre['preditions']
```

```python
df_pre['Differance (%)']=np.absolute(df_pre['Residual']/df_pre['Target']*100)
df_pre
```
